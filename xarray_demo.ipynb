### What kinds of data types does Xarray support?

Xarray has two datatypes: DataArrays and Datasets. 

DataArrays are designed to store a single variable. They are analogous to a single column or row of a spreadsheet, or a Pandas Series object. 

Datasets are designed to store many variables. In their simplest form, they are like spreadsheets or Pandas DataFrames. However, they extend much more easily to many dimensions than either of these.


### What the hell is a dimension?

Dimensions are basically independent variables. A set of dimensions should be enough to completely describe your data.

Let's say you are doing ion exchange membrane experiments. In your experimental design, you have two types of  membranes. You are modifying the membranes using an oxidation process. The oxidation process can be modified by selecting one of two oxidants, using a stronger concentration of oxidant, or a longer contact time.

In this case, the dims are (membrane type, type of oxidant, concentration of oxidant, contact time of oxidant). If you are doing multiple repetitions of an experiment, you might also have another dimension, which is repetition: (membrane type, type of oxidant, concentration of oxidant, contact time of oxidant, repetition.)

To use my own research, I did a lot of experiments using sodium-oxyanion salts and DIPA. My dims were: (solvent, salt, temperature, mixing ratio, phase, repetition). So if I took a measurement of an organic phase's salt concentration, I would have to write down which repetition it was, that it was in the organic phase, what the feed ratio was, the temperature, the salt, and the solvent used. 

### How does this help me?

Of course if you are a spreadsheet wizard, there are ways of dealing with all of this. However, spreadsheets are fundamentally 2-dimensional. As you go to higher dimensions, it gets easier (in my opinion) to revert to using an object that is designed for high-dimensional work.

There is a time cost to learn Xarray, but there would also be a time cost to learn Excel. I find Xarray satisfying and powerful to use, and hopefully I can convince you via this presentation that it's worth learning.

### This is pretty lame. Can't we do something more advanced?

Sure! What if we wanted to log pKw, the ionization constant of water, as a function of temperature?

```python
T_list = [0, 25, 50, 75, 100, 150, 200, 250, 300]
pKw_list = [14.95, 13.99, 13.26, 12.70, 12.25, 11.64, 11.31, 11.20, 11.34]
pKw = xr.DataArray(pKw_list, dims = ['T'])
```

```python
pKw
```

### OK, *we* know from the name of the data array that this is a list of pKws, but other people might not.

Good point. We definitely might want to know what this is describing. We can add metadata (information about the data) using attributes. Attributes are stored under

```python
pKw.attrs
```

As you can see, currently there are no attributes assigned to this DataArray. We can add attributes using dictionary-like notation:

```python
pKw.attrs["long name"] = "Negative log ionization constant of water"
pKw.attrs["short name"] = "pKw"
pKw.attrs["units"] = "–"
```

Now when we print pKw, we can see the attributes.

```python
pKw
```

### OK. This is a little better. But how will other people know that temperatures these correspond to? Without T_list, which is stored elsewhere, it's kind of useless. 

For sure! 

We can assign T_list as a "coord" or coordinate. Each coordinate corresponds to a dimension. Think of the dim like the x-axis and the coords as the actual values or ticks on the x-axis.

```python
pKw = pKw.assign_coords({'T':T_list})
```

```python
pKw
```

Now when we print pKw, each value in the DataArray has a temperature associated with it.

A lot of the time, it's better to put in the coords when we create the DataArray. We could have done this by writing

```python
pKw = xr.DataArray(pKw_list, dims = ['T'], coords = {'T':T_list})
```

Notice that when we do this, we pass the dims as a list of strings and coords as dictionary, where the keys are the dims.

```python
pKw
```

You can also assign extra coords to one dim, for example if you have different units of measure. Let's put everything into Kelvin and attach that as an extra set of coords on the T dimension.

```python
T_list_K = [T + 273.15 for T in T_list]
pKw = pKw.assign_coords(T_K = ('T',T_list_K))
```

```python
pKw
```

### I want to know what the pKw of water is at 50 C. Prof. Yip has given us a problem at that temperature.

To do that, you can use the sel function.

```python
pKw.sel({'T':50})
```

### What about at 85 C?

We don't have a value for 85 C in the DataArray. There are some built-in interpolation methods (https://docs.xarray.dev/en/stable/user-guide/interpolation.html?highlight=interpolation#interpolation-methods)
but I won't go into them.

But I can return the closest value, again using sel. This time it won't look for the exact temperature match, just the closest:

```python
pKw.sel({'T': 85}, method = 'nearest')
```

### Can we do some 2-D data?

Absolutely. Data that is 2-D but describing the same variable is best housed in a 2-D DataArray. For example, let's extend the pKw data into different pressures.

First, we have to load and clean the data. This is the most difficult part.

First, we can read in the data to a Pandas DataFrame.

```python
pKw_df = pd.read_excel('pKw.xlsx')
```

```python
pKw_df
```

There is a built-in function for converting DataFrames to xarray objects: df.to_xarray(). Let's see if that will do the trick for us.

```python
pKw_2d = pKw_df.to_xarray()
```

```python
pKw_2d
```

But this doesn't give us what we want. The dimension is (17,) instead of (17, 9). In other words, instead of having a 17 x 9 table of pKw values at 17 pressures and 9 temperatures, it's storing these pKw values as 9 separate DataArrays.

It's also storing the pressure as a variable, not a coord, and isn't even recognizing the temperature as a dimension. 

So in this case, to_xarray() isn't the right move (we will see in a little while how to use it properly).

### OK, can you show me how to properly process this table?

I can show you *my own* solution. There might be more efficient ones. 

First, I create numpy arrays for the pressure and temperature.

I extract the temperature by getting the columns of the dataframe.

```python
T = pKw_df.columns
```

```python
T
```

Currently, it is a numpy array of dtype 'object'. This is because we are mixing strings with ints. I don't want the 'Pressure (MPa)' thing out front, so I discard the first term, then typecast the data as floats.

```python
T = np.array(T[1:], dtype = 'float')
```

```python
T
```

The pressure is a little easier. I just reference the DataFrame column and do .values to make it into a numpy array.

```python
p = pKw_df['Pressure (MPa)'].values
```

```python
p
```

Finally, the data can be accessed using Pandas indexing. I want every row and all but the leftmost column (the leftmost column contains pressures, not pKw data).

```python
data = pKw_df.iloc[:, 1:]
```

```python
data
```

Now we can create our DataArray.

```python
pKw = xr.DataArray(data, dims = ['T', 'p'], coords = {'T':T, 'p': p})
```

Whoops, looks like we needed to put p before T in the dims list to match the shape of the array.

```python
pKw = xr.DataArray(data, dims = ['p', 'T'], coords = {'T':T, 'p': p})
```

```python
pKw
```

Let's double-check that everything looks good by selecting the pKw at 25 C and 0.1 MPa.

```python
pKw.sel({'T':25, 'p':0}, method = 'nearest')
```

Looks like we ran into a problem since the pressure is not monotonic -- I hadn't sorted by pressure previously!

No problem -- we can sort our DataArray easily:

```python
pKw = pKw.sortby(['p','T'])
```

```python
pKw
```

Now we can select.

```python
pKw.sel({'T': 25, 'p': 0}, method = 'nearest')
```

### Can we plot it?

Sure. Of course you can plot all the data using matplotlib or seaborn like you normally might, but there are a number of plotting options built into xarray (these are built on top of matplotlib).

```python
pKw.plot()
```

### What if I want to store multiple variables? 

Great question. This brings us to the second Xarray data type, the Dataset.

Datasets are used to store data that describe different things or have different units. For example, you might have some TSSE data where you have $x_s$, $x_a$, and $x_w$ for two different phases.

Let's load some data to see how we can store it efficiently. This is Eliza's NaBr/DIPA/water equilibrium data at 25 C.

```python
nabr_df = pd.read_csv('nabr_2.csv')
```

```python
nabr_df
```

OK. We need to figure out how many dims we have here. 













How many do we have?

Four altogether.

We have two dims, salt and temperature, which have only one coord ('NaBr' and 25.0, respectively).

We also have 'sample', which is really a proxy for mixing ratio or something like that.

Finally, we have phase. All of them except phase are in their own columns. 

If your data is arranged with each dim having its own column, then the import will go **seamlessly**.

We don't have that situation currently.

We could remedy this in Xarray. But we could equally do it in Excel.

```python
nabr_df = pd.read_csv('nabr.csv')
```

```python
nabr_df
```

Now we have a situation where every dim has a column, and every row has a coord in those dim columns.

We need to designate which columns contain dims. We do that by using the Pandas function set_index().

```python
nabr_df = nabr_df.set_index(['salt', 'T', 'phase', 'sample'])
```

Now nabr_df is a Pandas MultiIndex object.

```python
nabr_df
```

If we import it into xarray, things will go swimmingly:

```python
nabr = nabr_df.to_xarray()
```

```python
nabr
```

For some reason, there are three data variables that are full of nans. Let's just remove them using the .drop() function.

```python
nabr.drop(['Unnamed: 7','Unnamed: 8','Unnamed: 9'])
```

We can now take a look at our Dataset. It contains three variables, x_w, x_s, and x_a. If I want to access one of them, I can access it using dictionary-like syntax:

```python
nabr['x_a']
```

Notice that the x_a data is a DataArray. The individual variables in a Dataset are stored as DataArrays.

Let's define some attributes to go with this. For example, the solubility limit of nabr (as a mole fraction). Since we are defining this as an attribute, it is tied to all values in the Dataset.

```python
nabr.attrs['x_sat'] = 0.1421
```

We can easily define a new variable using dictionary-like syntax. We want to define a reduced solubility in the aqueous phase only.

```python
nabr['x_s_reduced'] = nabr['x_s'].sel({'phase':'a'}) / nabr.attrs['x_sat']
```

You will notice that x_s_reduced is not defined along the 'phase' dim. This is because we only applied it to the aqueous phase.

```python
nabr
```

OK, let's make our plot. We want our x to be x_s_red and our y to be x_s in the organic phase.

```python
fig, ax = plt.subplots()
x = nabr['x_s_reduced']
y = nabr['x_s']
ax.plot(x,y)
```

Whoops! We can only plot up to 2D data in matplotlib, and we only want to plot 1D data in any case.

We forgot to reduce the data down from its high dimensional state.

```python
x = nabr['x_s_reduced'].sel({'T': 25, 'salt':'nabr'})
y = nabr['x_s'].sel({'T': 25, 'salt':'nabr', 'phase':'o'})
```

Now x and y are 1D, and we are good to go.

```python
fig, ax = plt.subplots()
ax.plot(x,y, 'o')
```

### Pretty great. What about computing basic stats?

There are nifty computations that can be done automatically by xarray. Let's say we want the mean and standard deviation of x_s in the organic phase.

```python
ymean = y.mean(dim = 'sample')
ystd = y.std(dim = 'sample')
```

```python
ymean, ystd
```

Notice that these guys returned an Xarray DataArray object. The advantage is that this keeps the metadata, like the coords. But most of the time we just need the value. We can access that by using .values, which will return numpy arrays of just the data, without any coords or attrs.

```python
ymean.values, ystd.values
```

To use another example, you can easily find the maximum.

```python
y.max(), y.argmax(), y.idxmax() #Return the max value, the position of the max, and the coord of the max value.
```
